# **Project with class**
1- Library Management System :

```
# Book class
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def __str__(self):
        return f"'{self.title}' by {self.author} (Borrowed: {self.is_borrowed})"

# Member class
class Member:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if not book.is_borrowed:
            book.is_borrowed = True
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed {book.title}.")
        else:
            print(f"Sorry, '{book.title}' is already borrowed.")

    def return_book(self, book):
        if book in self.borrowed_books:
            book.is_borrowed = False
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title}.")
        else:
            print(f"{self.name} does not have '{book.title}' borrowed.")

# Library class
class Library:
    def __init__(self):
        self.books = []
        self.members = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added {book.title} to the library.")

    def add_member(self, member):
        self.members.append(member)
        print(f"Added member: {member.name}.")

    def show_books(self):
        print("\nAvailable Books:")
        for book in self.books:
            print(book)

# Example usage
if __name__ == "__main__":
    # Create a library
    library = Library()

    # Add books to the library
    book1 = Book("1984", "George Orwell")
    book2 = Book("To Kill a Mockingbird", "Harper Lee")
    book3 = Book("The Great Gatsby", "F. Scott Fitzgerald")

    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)

    # Add members to the library
    member1 = Member("Alice")
    member2 = Member("Bob")

    library.add_member(member1)
    library.add_member(member2)

    # Show available books
    library.show_books()

    # Members borrow books
    member1.borrow_book(book1)
    member1.borrow_book(book2)

    # Attempt to borrow an already borrowed book
    member2.borrow_book(book1)

    # Show available books after borrowing
    library.show_books()

    # Members return books
    member1.return_book(book1)

    # Show available books after returning
    library.show_books()
```

2-Simple Banking System

```
# Account class
class Account:
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}.")
        elif amount > self.balance:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        return self.balance

# Customer class
class Customer:
    def __init__(self, name):
        self.name = name
        self.accounts = []

    def create_account(self, account_number):
        new_account = Account(account_number, self.name)
        self.accounts.append(new_account)
        print(f"Account {account_number} created for {self.name}.")

    def get_accounts(self):
        return self.accounts

# Bank class
class Bank:
    def __init__(self):
        self.customers = []

    def add_customer(self, customer):
        self.customers.append(customer)
        print(f"Added customer: {customer.name}.")

    def show_customers(self):
        print("\nCustomers:")
        for customer in self.customers:
            print(customer.name)

# Example usage
if __name__ == "__main__":
    # Create a bank
    bank = Bank()

    # Create customers
    customer1 = Customer("Alice")
    customer2 = Customer("Bob")

    # Add customers to the bank
    bank.add_customer(customer1)
    bank.add_customer(customer2)

    # Create accounts for customers
    customer1.create_account("ACC1001")
    customer1.create_account("ACC1002")

    customer2.create_account("ACC2001")

    # Perform transactions on accounts
    account1 = customer1.get_accounts()[0]
    account1.deposit(500)
    account1.withdraw(200)

    account2 = customer1.get_accounts()[1]
    account2.deposit(300)

    account3 = customer2.get_accounts()[0]
    account3.deposit(1000)
    account3.withdraw(1500)  # Insufficient funds

    # Show customers and their accounts
    bank.show_customers()

    for customer in bank.customers:
        print(f"\nAccounts for {customer.name}:")
        for account in customer.get_accounts():
            print(f"Account Number: {account.account_number}, Balance: ${account.get_balance():.2f}")
```
