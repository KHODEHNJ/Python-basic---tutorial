# Python data types :

In python we have some data types that all of them has their specefic feature

# 1- **String**
These types are like text and we write it like this
'Hello world' or "Hello world"

All string cant use like numbers and their operation

Example :
`print("Hello world")`

# 2- **Int**
Int is all number like this ... , -1 , 0 , , 1 , ...
not like 4.0 , -2.032 ...

Example :
`print(2)`

# 3- **Float**
float is all numbers like this 2.0 , 3.0931

Example :
`print(2.0)`

most important thing about numbers is :
if we write "2" or "3.932" **these all are string not int or float**
